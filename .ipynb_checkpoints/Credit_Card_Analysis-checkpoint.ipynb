```python
import pandas as pd
from sklearn.model_selection import learning_curve, train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from matplotlib import pyplot as plt
import seaborn as sns
import xgboost as xgb
import statsmodels.api as sm
```

```python
data = pd.read_csv('./UCI_Credit_Card.csv')
pd.set_option('display.max_columns', None)
data.drop('ID', axis = 1, inplace = True)
print(data.head())
```

```python
print(data.shape)
```

```python
print(data.describe())
```

```python
# Check the next month status
next_month = data['default.payment.next.month'].value_counts()
print(next_month)
df = pd.DataFrame({'default.payment.next.month': next_month.index,'values': next_month.values})
plt.figure(figsize = (6,6))
plt.title('Payment Overdue\n (Good: 0, Overdue: 1)')
sns.set_color_codes("pastel")
sns.barplot(x = 'default.payment.next.month', y="values", data=df)
locs, labels = plt.xticks()
plt.show()
```

```python
# 特征选择，去掉ID字段、最后一个结果字段即可
target = data['default.payment.next.month'].values
columns = data.columns.tolist()
columns.remove('default.payment.next.month')
features = data[columns].values
```

```python
corr = data[columns].corr()
plt.figure(figsize=(23,23))
# annot=True show all the blocks
# Some of the features may have extremely high correlations, anti-multi-colinearity experiment start
sns.heatmap(corr, annot=True)
plt.show()
```

```python
PAY = list(data.columns[5:11])
BILL = list(data.columns[11:17])
PAY_data = data[PAY]
BILL_data = data[BILL]
print(PAY_data.head())

for i in range(0, len(data.columns)):
    y = data.loc[:, data.columns == data.columns[i]]
    x = data.loc[:, data.columns != data.columns[i]]
    model = sm.OLS(y, x)
    results = model.fit()
    rsq = results.rsquared
    vif = round(1 / (1 - rsq), 2)
    print(
        "R Square value of {} column is {} keeping all other columns as features".format(
            data.columns[i], (round(rsq, 2))
        )
    )
    print(
        "Variance Inflation Factor of {} column is {} \n".format(
            data.columns[i], vif)
        )


```

```python
# Based on the vif, all the BILL amt will be removed for experiment
data.drop(['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', "BILL_AMT4", 'BILL_AMT5', 'BILL_AMT6'], inplace=True, axis =1)
columns_exp = data.columns.tolist()

features_exp = data[columns_exp].values
```

```python

corr_exp = data[columns_exp].corr()
plt.figure(figsize=(17,17))
# annot=True show all the blocks
# Some of the features may have extremely high correlations, anti-multi-colinearity experiment start
sns.heatmap(corr_exp, annot=True)
plt.show()
```

```python
# 30%作为测试集，其余作为训练集
train_x, test_x, train_y, test_y = train_test_split(features, target, test_size=0.30, stratify = target, random_state = 1)
    
# 构造各种分类器
classifiers = [
    SVC(random_state = 1, kernel = 'rbf'),    
    DecisionTreeClassifier(random_state = 1, criterion = 'gini'),
    RandomForestClassifier(random_state = 1, criterion = 'gini'),
    KNeighborsClassifier(metric = 'minkowski'),
    xgb.XGBClassifier()
]
# 分类器名称
classifier_names = [
            'svc', 
            'decisiontreeclassifier',
            'randomforestclassifier',
            'kneighborsclassifier',
            'XGBoost',
]

# 分类器参数
classifier_param_grid = [
            {'svc__C':[1], 'svc__gamma':[0.01]},
            {'decisiontreeclassifier__max_depth':[6,9,11]},
            {'randomforestclassifier__n_estimators':[3,5,6]} ,
            {'kneighborsclassifier__n_neighbors':[4,6,8]},
            {'XGBoost__max_depth':[3, 4, 5], 'XGBoost__n_estimators':[100, 150, 200]},
]
 
# 对具体的分类器进行GridSearchCV参数调优
def GridSearchCV_work(pipeline, train_x, train_y, test_x, test_y, param_grid, score = 'accuracy'):
    response = {}
    gridsearch = GridSearchCV(estimator = pipeline, param_grid = param_grid, scoring = score)
    # 寻找最优的参数 和最优的准确率分数
    search = gridsearch.fit(train_x, train_y)
    print("GridSearch最优参数：", search.best_params_)
    print("GridSearch最优分数： %0.4lf" %search.best_score_)
    predict_y = gridsearch.predict(test_x)
    print("准确率 %0.4lf" %accuracy_score(test_y, predict_y))
    response['predict_y'] = predict_y
    response['accuracy_score'] = accuracy_score(test_y,predict_y)
    return response
 
for model, model_name, model_param_grid in zip(classifiers, classifier_names, classifier_param_grid):
    pipeline = Pipeline([
            ('scaler', StandardScaler()),
            (model_name, model)
    ])
    result = GridSearchCV_work(pipeline, train_x, train_y, test_x, test_y, model_param_grid , score = 'accuracy')
```
